```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, BatchNormalization, Dropout, Dense
from deep_translator import GoogleTranslator
from sklearn.model_selection import train_test_split

# Set API key for Langsmith (if needed)
os.environ['LANGSMITH_API_KEY'] = 'your_langsmith_api_key'

# Initialize the language classification model
data = pd.read_csv('/Users/damacm1143/Downloads/New Folder With Items/SA_official_languages.csv')
texts = data['text'].values
labels = data['lang_id'].values

# Tokenize and pad sequences for text classification
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
word_index = tokenizer.word_index

maxlen = 100   
data = pad_sequences(sequences, maxlen=maxlen)

label_tokenizer = Tokenizer()
label_tokenizer.fit_on_texts(labels)
label_sequences = np.array(label_tokenizer.texts_to_sequences(labels)) - 1

X_train, X_test, y_train, y_test = train_test_split(data, label_sequences, test_size=0.2, random_state=42)

# Define and train the language classification model
model = Sequential([
    Embedding(input_dim=len(word_index) + 1, output_dim=128, input_length=maxlen),
    Bidirectional(LSTM(64, return_sequences=True)),
    BatchNormalization(),
    Dropout(0.5),
    Bidirectional(LSTM(64)),
    BatchNormalization(),
    Dropout(0.5),
    Dense(len(label_tokenizer.word_index), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=50, batch_size=40, validation_split=0.2)

# Function to predict the language of a given text
def predict_language(text):
    sequence = tokenizer.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=maxlen)
    prediction = model.predict(padded_sequence)
    label_index = np.argmax(prediction) + 1
    label = label_tokenizer.index_word.get(label_index, 'unknown')
    return label

# Function to translate text to English
def translate_to_english(text, source_language):
    try:
        translator = GoogleTranslator(source=source_language, target='en')
        translated_text = translator.translate(text)
        if not translated_text:
            raise ValueError("Received None or incomplete result from the translation service.")
        return translated_text
    except Exception as e:
        print(f"An error occurred during translation: {e}")
        return None

# Example usage
def process_and_translate(text):
    try:
        # Predict the language of the input text
        lang_id = predict_language(text)
        print(f"Detected language: {lang_id}")

        # Translate the text to English if it's not already in English
        if lang_id != 'en':
            translated_text = translate_to_english(text, source_language=lang_id)
            if translated_text is not None:
                print(f"Translated to English: {translated_text}")
            else:
                print(f"Failed to translate text for language_id: {lang_id}")
        else:
            print(f"Text is already in English: {text}")

    except Exception as e:
        print(f"Processing error: {e}")

# Example prediction
process_and_translate("Ndi nga imisa hani oda ya debit?")
```

```python
import os
from langchain.llms import Ollama
from langchain.embeddings import OllamaEmbeddings
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import create_retrieval_chain
from langchain import hub
from langchain.chains.combine_documents import create_stuff_documents_chain
from deep_translator import GoogleTranslator
import langid

# Set the API key
os.environ['LANGSMITH_API_KEY'] = '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'

# Ensure consistent language detection
langid.set_languages(['en', 'zu', 'xh', 'af'])

# Initialize the models
llm = Ollama(model="llama3.1", base_url="http://127.0.0.1:11434")
embed_model = OllamaEmbeddings(model="llama3.1", base_url='http://127.0.0.1:11434')

def translate_text(text, target_language='en'):
    try:
        # Perform the translation using deep-translator's GoogleTranslator
        translated_text = GoogleTranslator(source='auto', target=target_language).translate(text)
        if not translated_text:
            raise ValueError("Received None or incomplete result from the translation service.")
        return translated_text
    except Exception as e:
        print(f"An error occurred during translation: {e}")
        return None

def process_text(text):
    try:
        # Detect the language of the input text using langid
        lang_id, _ = langid.classify(text)
        print(f"Detected language: {lang_id}")

        # Translate the text to English if it's not already in English
        if lang_id != 'en':
            translated_text = translate_text(text, target_language='en')
            if translated_text is not None:
                text = translated_text
            else:
                print(f"Failed to translate text for language_id: {lang_id}")
                return

        # Process the text using Ollama
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=128)
        chunks = text_splitter.split_text(text)
        vector_store = Chroma.from_texts(chunks, embed_model)
        retriever = vector_store.as_retriever()
        retrieval_qa_chat_prompt = hub.pull("langchain-ai/retrieval-qa-chat")
        combine_docs_chain = create_stuff_documents_chain(llm, retrieval_qa_chat_prompt)
        retrieval_chain = create_retrieval_chain(retriever, combine_docs_chain)

        # Retrieve information from the text
        response = retrieval_chain.invoke({"input": "President Cyril Ramaphosa has issued a new bill of NHI and a two-pot retirement system"})
        answer = response.get('answer', "No answer found")
        print(f"Answer: {answer}")

        # Translate the response back to the original language (if needed)
        if lang_id != 'en':
            response_translation = translate_text(answer, target_language=lang_id)
            if response_translation is not None:
                print(f"Translation to {lang_id}: {response_translation}")
            else:
                print(f"Failed to translate the response to {lang_id}")
        else:
            print(f"Response: {answer}")

    except Exception as e:
        print(f"Processing error: {e}")

# Example usage
your_text = "Ngubani umongameli woMzantsi Afrika?"
process_text(your_text)  # Detect language and process
```

```python
import os
from langchain_community.llms import Ollama
from dotenv import load_dotenv
from langchain_community.embeddings import OllamaEmbeddings
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain

# Define the prompt directly in the script
retrieval_qa_chat_prompt = "Please provide a detailed answer based on the provided documents."

# Initialize the LLM
llm = Ollama(model="llama3.1", base_url="http://127.0.0.1:11434")

# Initialize the embedding model
embed_model = OllamaEmbeddings(model="llama3.1", base_url="http://127.0.0.1:11434")

# Prepare the text
text = """
    ke nyaka go ya go bona lešole la mosadi go tšwa Jeremane
    """

# Split the text into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=128)
chunks = text_splitter.split_text(text)

# Create the vector store
vector_store = Chroma.from_texts(chunks, embed_model)

# Create the retriever
retriever = vector_store.as_retriever()

# Create the retrieval chain
combine_docs_chain = create_stuff_documents_chain(llm, retrieval_qa_chat_prompt)
retrieval_chain = create_retrieval_chain(retriever=retriever, combine_docs_chain=combine_docs_chain)

# Set the API key if necessary (Ensure it's needed for your setup)
os.environ['LANGSMITH_API_KEY'] = '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'

# Invoke the chain with a query
response = retrieval_chain.invoke({"input": "ke mang motho o ba bolelang ka yena?"})

# Print the response
print(response['answer'])
```
